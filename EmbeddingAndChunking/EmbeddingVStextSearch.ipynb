```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Initialize the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sample documents
documents = [
    "The cat sat on the mat.",
    "Dogs are great pets.",
    "I love programming in Python.",
    "The quick brown fox jumps over the lazy dog.",
    "Artificial Intelligence is transforming the world."
]

# Convert documents to embeddings
document_embeddings = model.encode(documents)
```

```python
# Convert query to embedding
query = "AI is transforming the world."
query_embedding = model.encode(query)
```

```python
# Calculate cosine similarity between query and document embeddings
similarities = cosine_similarity([query_embedding], document_embeddings)[0]

# Find the most similar document
most_similar_index = np.argmax(similarities)
most_similar_document = documents[most_similar_index]

print("Query:", query)
print("Most similar document:", most_similar_document)
```

```python
# Define a simple text search function
def text_search(query, documents):
    results = []
    for doc in documents:
        if query.lower() in doc.lower():
            results.append(doc)
    return results

# Perform text search
text_search_results = text_search(query, documents)

print("Text Search Results:", text_search_results)
```
